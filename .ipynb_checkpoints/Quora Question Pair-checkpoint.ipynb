# Approach 1 : Fit the Bag of Words into Random Forest

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import spacy
```

```python
df = pd.read_csv("questions.csv")
df.head(10)
```

```python
qid = pd.Series(df.question1.tolist() + df.question2.tolist())
plt.hist(qid.value_counts().values, bins=160)
plt.xlim([0, 20])
plt.show()
```

```python
df = df.sample(3000, random_state=2)
```

```python
df.isnull().sum()
```

```python
df.head(10)
```

```python
ques_df = df.iloc[:,[3,4]]
ques_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

questions = list(ques_df.question1) + list(ques_df.question2)

cv = CountVectorizer(max_features = 3000)
q1_arr, q2_arr = np.split(cv.fit_transform(questions).toarray(), 2)
```

```python

temp_df1 = pd.DataFrame(q1_arr, index = ques_df.index)
temp_df2 = pd.DataFrame(q2_arr, index = ques_df.index)
temp_df = pd.concat([temp_df1, temp_df2], axis=1)
temp_df.shape
```

```python
temp_df["is_duplicate"] = df.is_duplicate
temp_df
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(temp_df.iloc[:,0:-1].values, temp_df.iloc[:,-1].values, test_size = 0.2, random_state = 1)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

rf = RandomForestClassifier()
rf.fit(X_train, y_train)
```

```python
y_predict = rf.predict(X_test)
accuracy_score(y_test, y_predict)
```

## Fitting into Extreme Gradient Boosting

```python
from xgboost import XGBClassifier
xgb = XGBClassifier()
xgb.fit(X_train, y_train)
```

```python
y_predict = xgb.predict(X_test)
accuracy_score(y_test, y_predict)
```

# Approach 2 : Let's introduce some features
- __q1_len__ = char length of question 1
- __q2_len__ = char length of question 2
- __words_common__ = number of common words in both questions
- __words_total__ = number of total words in both questions
- __words_share__ = words_common / words_total

```python
df["q1_len"] = df.question1.str.len()
df["q2_len"] = df.question2.str.len()
df.head()
```

```python
df["q1_words_total"] = df.question1.apply(lambda row: len(row.split(" ")))
df["q2_words_total"] = df.question2.apply(lambda row: len(row.split(" ")))
df.head()
```

```python
def common_words(row):
    w1 = set(map(lambda word: word.lower().strip(), row.question1.split(" ")))
    w2 = set(map(lambda word: word.lower().strip(), row.question2.split(" ")))
    return len(w1&w2)
```

```python
df["words_common"] = df.apply(common_words, axis=1)
df.head()
```

```python
def total_words(row):
    w1 = set(map(lambda word: word.lower().strip(), row.question1.split(" ")))
    w2 = set(map(lambda word: word.lower().strip(), row.question2.split(" ")))
    return (len(w1) + len(w2))
```

```python
df["words_total"] = df.apply(total_words, axis=1)
df.head()
```

```python
df["word_share"] = round(df.words_common/df.words_total , 2)
df.head()
```
